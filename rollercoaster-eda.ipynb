```python
#Asking a Question about the Data
#mean speed of roller coasters by location.
```

```python
ax = (
    df.query('Location != "Other"')
      .groupby('Location')['Speed MPH']
      .agg(['mean', 'count'])
      .query('count >= 10')
      .sort_values('mean')
      ['mean']
      .plot(kind='barh', figsize=(12, 8), title='Average Coaster Speed by Location')
)

ax.set_xlabel('Average Coaster Speed')
```
